```python
import requests
import json
from datetime import datetime
import time

def grab_player_stats (username):
    url = f'https://jstris.jezevec10.com/u/{username}/stats?mode=1&displayAll=true'
    page = requests.get(url).text
    data = json.loads(page.split('var data =')[1].split('</script>')[0][2:-7]) 

    with open(f'jstris_data-{username}.tsv','w') as f:
        f.write('Time\tReplay\tPlayed\n')
        for i in data['skyline'] + data['unused']:
            f.write(str(i['y'])+'\t'+str(i['id'])+'\t'+datetime.utcfromtimestamp(i['x']).strftime('%Y-%m-%d %H:%M:%S')+'\n')
    

grab_player_stats ('Not_Kaiyon') 
```

```python
import pandas as pd
df = pd.read_csv (r"jstris_data-NotBBG.tsv", sep="\t")
df.columns = df.columns.str.lower()
df.dropna(inplace=True)
df.head()

df["propper_date"] = pd.to_datetime(df["played"])
df["dayssincestart"]= (df["propper_date"] - df["propper_date"].iloc[0]).dt.total_seconds() /(24 *3600)
df
```
