```python
import pandas as pd
import json
# x = {"a": {"a1": 1, "a2":2}, "b": {"a1": 3, "a2": 4}}
with open (r"C:\Users\micha\Downloads\Bens_Tetris_Regression\regression_data.json", "r") as f:
    x = json.load(f)
    
x = {key: {**{k:val[k] for k in val if k != "metrics"}, **val["metrics"]} for key, val in x.items() }
x = {key: {**{k:val[k] for k in val if k != "params"}, **{p_name: p_val for p_val, p_name in zip(val["params"], ['a', 'b', 'c'])}} for key, val in x.items() }

df = pd.DataFrame(x).T
df.index.name = "playername"
df = df.loc[df["gamesplayed"] > 200]
# df.drop(columns=['warnings'], inplace = True)
df
```

```python
df.to_csv ('regression_data.csv', index = True, header=True)
```
